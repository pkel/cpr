```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("../../data/bounds.csv")
plt = sns.relplot(
    data=df,
    kind="line",
    hue="model",
    x="k",
    y="p",
    facet_kws=dict(sharey=False),
    row="alpha",
    col="block_interval",
)
plt.set(yscale="log")
```
